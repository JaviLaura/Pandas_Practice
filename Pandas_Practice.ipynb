```python
import pandas as pd
#Filtering
csv1 = pd.read_csv('https://static.junilearning.com/python_data_science/tips.csv')
display(csv1)
```

```python
filteredData = csv1[csv1['day'] == 'Sun']
filteredData.describe()
```

```python
print(csv1) #What is the difference between printing and displaying a data frame? Printing means printing text, but display is better for graphics.
```

```python
filteredData = csv1[csv1['tip'] < 2]
display(filteredData)
```

```python
filteredData = csv1[(csv1['day'] == 'Fri') & (csv1['sex'] == 'Female')] 
display(filteredData)
```

```python
filteredData = csv1[((csv1['day'] == 'Fri') | (csv1['day'] == 'Thur')) & (csv1['tip'] == 3)]   #| is or. 
display(filteredData)
```

```python
filteredData = csv1[(csv1['day'] == 'Fri') | (csv1['day'] == 'Thur')]
display(filteredData)
```
